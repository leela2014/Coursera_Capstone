```python
!pip install bs4
import numpy as np 
import pandas as pd 
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
source = requests.get(url).text
can_html  = BeautifulSoup(source, 'xml')
can_html = can_html.find('table')
```

```python
row_data = []
for tr_cell in can_html.find_all('tr'):
    for td_cell in tr_cell.find_all('td'):
        row_data.append(td_cell.text.strip())
        #print(td_cell.text.strip())
print(len(row_data))
row_data[:10]
```

```python
#extract data into a list
borough  = [row_data[i][3:] for i in range(len(row_data))]
postalCode = [row_data[i][:3] for i in range(len(row_data))]
print(postalCode[:5])
print(borough[:5])
```

```python
# make df 
can_df = pd.DataFrame()
can_df['Postalcode'] = postalCode
can_df['temp'] = borough
can_df.head()
```

```python
can_df = can_df[can_df.temp != "Not assigned"] # delect df row which are 'Not assigned'
can_df[['Borough','Neighborhood','na']] = can_df.temp.str.split("(",expand = True,) # split col:temp
can_df['Neighborhood'] = can_df['Neighborhood'].str.replace(')','') 
can_df['Neighborhood'] = can_df['Neighborhood'].str.replace('/',',')
can_df = can_df.drop(['temp','na'],axis=1) # drop columns
can_df.head()
```

```python
print(can_df.shape)
```

```python
#import lat long data from csv file
latlong = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DS0701EN-SkillsNetwork/labs_v1/Geospatial_Coordinates.csv')

latlong.head(10)
```

```python
#join dataframes
can_df = can_df.join(latlong.set_index('Postal Code'), on='Postalcode')
can_df.head(10)
```

*** Visualization: Tornoto map

```python
# Create Toronto map
map_toronto = folium.Map(location=[43.65, -79.38], zoom_start=10)

# Add markers
for lat, lng, borough, neighborhood in zip(can_df['Latitude'], can_df['Longitude'], can_df['Borough'], can_df['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=7,
        popup=label,
        color='green',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```

*** K clusterning ***

```python
k=5
toronto_clustering = can_df.drop(['Postalcode','Borough','Neighborhood'],axis=1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
can_df.insert(0, 'Cluster_Labels', kmeans.labels_)
```

```python
can_df.head()
```

```python
can_df['Cluster_Labels'].value_counts()
```

```python
# create map
map_clusters = folium.Map(location=[43.65, -79.38],zoom_start=10)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(can_df['Latitude'], can_df['Longitude'], can_df['Neighborhood'], can_df['Cluster_Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
